# Ensemble Learning

### 1. What is Ensemble Learning?

Ensemble Learning is a technique in machine learning where multiple models (often called "weak learners") are combined to produce a more accurate and robust model (strong learner) than any individual model.

ðŸ‘‰ The basic idea:
"A group of weak models can perform better than a single strong model if combined properly."

This is similar to how a committee of experts (ensemble) can make better decisions than one expert alone.

### ðŸ”¹ 2. Why Ensemble Learning?
- A single model may suffer from bias (underfitting) or variance (overfitting).
- Combining models helps in reducing errors.
- Works especially well in classification and regression tasks.

Mathematically, if models are independent and have error rate Îµ < 0.5, then combining multiple models reduces the overall error probability drastically.

### ðŸ”¹ 3. Types of Ensemble Learning

Ensemble methods are broadly divided into two categories:

##### A. Sequential Ensembles
- Models are built one after another.
- Each model tries to reduce errors made by the previous one.

Example: Boosting

##### B. Parallel Ensembles
- Models are built independently in parallel.
- Final prediction is made by combining their outputs (like voting or averaging).

Example: Bagging, Random Forest
