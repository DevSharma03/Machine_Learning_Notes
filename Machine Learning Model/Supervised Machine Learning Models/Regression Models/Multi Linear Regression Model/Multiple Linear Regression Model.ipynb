# Multiple Linear Regression

### What is Multiple Linear Regression?
- Multiple Linear Regression models the linear relationship between one continuous dependent variable and two or more independent variables.
- Multiple Linear Regression is an extension of simple linear regression as it takes more than one predictor varible to predict the responce variable.
- We will use the multiple linear Regression model when we see the linearity in between the 2 or more feature with the target variable.
- Here, a plane is created at the place of plane.

###  Python Code Example (with scikit-learn) By ChatGPT

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Inputs (2 features): [Hours Studied, Sleep Hours]
X = np.array([
    [1, 8],
    [2, 7],
    [3, 6],
    [4, 5],
    [5, 4]
])
```

```python
# Output: Exam Scores
y = np.array([50, 55, 60, 65, 70])
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Coefficients
print("Intercept (w0):", model.intercept_)
print("Coefficients (w1, w2):", model.coef_)

# Predict score for 6 hours study, 5 hours sleep
predicted = model.predict([[6, 5]])
print("Predicted Exam Score:", predicted[0])
```

###  Python Code Example (with scikit-learn) By YouTube

```python
# Import Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Multiple Linear Regression Data Set.csv")
ds.head()
```

```python
ds.shape, ds.isnull().sum()
```

```python
sns.pairplot(data = ds)  # use to check the co-linearity of the data between each other
plt.show()
```

```python
sns.heatmap(data = ds.corr(), annot= True)  # use to check the co-linearity of the data between each other
plt.show()
```

```python
x = ds.drop('income', axis = 1)
y = ds['income']
x, y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state= 42)
x_train, x_test, y_train, y_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Coefficients
print("Intercept (w0):", model.intercept_)
print("Coefficients (w1, w2):", model.coef_)
model.predict([[30, 3]])
```

```python
model.score(x_test, y_test) * 100
```
