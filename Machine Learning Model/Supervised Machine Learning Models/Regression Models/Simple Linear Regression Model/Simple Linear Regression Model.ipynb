# Simple Linear Regression

### What is Simple Linear Regression?
- Simple Linear Regression is a Supervised Learning algorithm that predicts a continuous target variable using a single input feature by modeling a linear relationship between them.
- It is a type of regression algorithm that models the relationship between the dependent variable and a single independent variable.
- We use it when we only have one input variable and one output variable to work work with.

- Equation of a line:

   Y = mx + c

- Where:
- Y: predicted output
- x: input feature
- c: intercept (bias)
- m: slope (weight)/gradient/coefficent

- We aim to learn the best-fitting line through the training data that minimizes the error between predicted and actual values.

# Practical Implementation of Simple Linear Regression

##### Code Implementation (Python using scikit-learn) From ChatGPT

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([50, 55, 65, 70, 75])
```

```python
# Model
model = LinearRegression()
model.fit(X, y)
```

```python
# Prediction
print("Slope (w1):", model.coef_[0])
print("Intercept (w0):", model.intercept_)
print("Predicted score for 6 hours:", model.predict([[6]])[0])
```

```python
# Visualization
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Simple Linear Regression")
plt.show()
```

##### Code Implementation from Youtube

```python
# Import Libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
ds = pd.read_csv('Simple Linear Regression Dataset.csv')
ds.head()
```

```python
sns.scatterplot(x= 'x', y= 'y', data= ds)   # to check the linearity of the dataset which we are working on.
plt.show()
```

```python
ds.isnull().sum()
```

```python
# spliting the data into X, y (feature and targets)
X = ds[['x']]
y = ds['y']
X, y
```

```python
# split the dataset in train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.20, random_state=42)

# Print the data
X_train, X_test, y_train, y_test
```

```python
model = LinearRegression()
model.fit(X_train, y_train)   # the fit function here is used to train the model to get the best value of m & c in the eqation "y = mx + c"
```

```python
model.predict([[25]])
```

```python
# Check Accuracy
model.score(X_test, y_test) * 100   # This will provide the accuracy of the model based on test data
```

```python
model.coef_    # this is the value of m in the equation as Slope
```

```python
model.intercept_  # this is the value of C in the equation as intercept
```

```python
result = model.predict(X)

# Visualization
sns.scatterplot(x= 'x', y= 'y', data= ds)   # to check the linearity of the dataset which we are working on.
plt.plot(ds["x"], result, c= 'red')
plt.legend(['Org Data', 'Predicting Line'])
plt.show()
```
