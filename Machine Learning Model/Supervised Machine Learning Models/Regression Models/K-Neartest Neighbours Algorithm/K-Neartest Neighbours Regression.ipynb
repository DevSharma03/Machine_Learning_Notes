# K - Nearest Neighbor Algorithm

- K-NN algorithm can be used for Regression as well as for classification but mostly it is used for the classification problems.
- K-NN is a non-parametric algorithm, which means it does not make any assumption on underlying data.
- It is also called a lazy learner algorithm
- It works by looking at the nearest neighbors of the new data point which needs to be predicted and it checks K no. of nearest neighbors.

# K-Nearest Neighbors Regression

### What is K-NN Regression?
- K-NN Regression is a supervised learning algorithm that predicts a continuous target value by averaging the values of the k nearest neighbors in the feature space.
- Unlike K-NN Classification (which takes a vote), K-NN Regression uses mean (or weighted mean) of the k closest observations.

### Python Example – KNN Regressor From ChatGPT

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic dataset
np.random.seed(0)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.1 * np.random.randn(100)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train KNN regressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict
y_pred = knn.predict(X_test)

print(knn.score(X_test, y_test)*100)

# Evaluation
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))

# Visualization
X_plot = np.linspace(0, 5, 100).reshape(-1, 1)
y_plot = knn.predict(X_plot)

plt.scatter(X, y, label="Training data", color='gray')
plt.plot(X_plot, y_plot, label="KNN prediction", color='red')
plt.xlabel("X")
plt.ylabel("y")
plt.title("K-NN Regression")
plt.legend()
plt.grid(True)
plt.show()
```
